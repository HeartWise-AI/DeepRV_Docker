```python
import yaml
from orion.utils import video_run_inference
```

```python
def run_inference(output_folder, config_file_path = '/volume/DeepRV/config/config_classification_inference.yaml'):   
   with open(config_file_path) as file:
        config = yaml.safe_load(file)
        print(config['model_path'])        
        print(config['num_epochs'])
        df_predictions_inference = video_run_inference.run_inference_and_no_logging(
          config_path=config_file_path,
          model_path = config['model_path'],
          data_path = config['data_filename'],
          output_dir=output_folder,
          split='inference')
     

    # return predictions dataframe as per your need
   return df_predictions_inference
```
